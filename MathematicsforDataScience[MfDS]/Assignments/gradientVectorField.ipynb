```python
def gradient_function(x,y):
    """
    Calculates the gradient by partial derivatives of the function f(x,y) = x¬≤y - xy¬≤ at this point
    """
    # Since ‚àáùëì(ùë•, ùë¶) = ‚å©ùëìùë•(ùë•, ùë¶), ùëìùë¶(ùë•, ùë¶)‚å™ -> we will have ‚àáùëì(ùë•, ùë¶) = ‚å©2ùë•ùë¶ - ùë¶¬≤, ùë•¬≤ - 2ùë•ùë¶‚å™
    fx, fy = 2*x*y - y**2, x**2 - 2*x*y

    return fx, fy
```

```python
def main():
    """
    Main function that asks user inputs and gives the gradient of the function f(x,y) = x¬≤y - xy¬≤
    """
    try:
        x = float(input("Enter the value of x: "))
        y = float(input("Enter the value of y: "))
        print("The gradient of the function f(x,y) = x¬≤y - xy¬≤ at this point is: ", gradient_function(x, y))
    except ValueError:
        print("You have entered a non numerical value.")

main()
```

```python
"""
Gradient vector field graph with the help of numpy and matplotlib
"""
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

xline, yline = np.meshgrid(np.linspace(-10, 10, 15), np.linspace(-10, 10, 15))

fx = 2*xline*yline - yline**2
fy = xline**2 - 2*xline*yline

plt.quiver(xline, yline, fx, fy)
plt.show()
```
